```python
import re

dict = {}
sentence_list = ['This is a sentence', 'This is a sentence']
for sentence in sentence_list:
    for word in re.split('\s', sentence): # split with whitespace
        try:
            dict[word] += 1
        except KeyError:
            dict[word] = 1
print(dict)
```

```python
import re

dict = {}
sentence_list = ['This is a sentence', 'This is a sentence']
for sentence in data_train['question_text']:
    for word in re.split('\s', str(sentence)): # split with whitespace
        try:
            dict[word] += 1
        except KeyError:
            dict[word] = 1
#     print(1)
# print(dict.most_common())
```

```python
dict
```

```python
import operator
sorted_d = sorted(dict.items(), key=operator.itemgetter(1),reverse = True)
```

```python
sorted_d
```

```python
with open("count.txt",'w',encoding='utf-8') as f:
    for i in sorted_d:
        f.write(i)
```
